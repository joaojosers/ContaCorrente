```python
df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'localizacao',
    'company_size': 'porte'
}, inplace=True)
```

```python
print('Nomes das colunas após a tradução:')
display(df.columns)
display(df.head())
```

```python
df["remoto"].value_counts()
```

```python
df["porte"].value_counts()
```

```python
df['senioridade'] = df['senioridade'].map({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})

print("Valores da coluna 'senioridade' traduzidos:")
display(df['senioridade'].value_counts())
```

```python
substituir_senioridade = {
    'Senior': 'SE',
    'Pleno': 'MI',
    'Junior': 'EN',
    'Executivo': 'EX'
}

df['senioridade'] = df['senioridade'].map(substituir_senioridade)
df['senioridade'].value_counts()
```

```python
df['contrato'] = df['contrato'].map({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

print("Valores da coluna 'contrato' traduzidos:")
display(df['contrato'].value_counts())
```

```python
df['porte'] = df['porte'].map({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})

print("Valores da coluna 'porte' traduzidos:")
display(df['porte'].value_counts())
```

```python
print("Valores da coluna 'porte' traduzidos:")
display(df['porte'].value_counts())
```

```python
df['remoto'] = df['remoto'].map({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
})

print("Valores da coluna 'remoto' traduzidos:")
display(df['remoto'].value_counts())
```
