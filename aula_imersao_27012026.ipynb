```python
df['senioridade'] = df['senioridade'].map({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})

print("Valores da coluna 'senioridade' traduzidos:")
display(df['senioridade'].value_counts())
```

```python
df['contrato'] = df['contrato'].map({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

print("Valores da coluna 'contrato' traduzidos:")
display(df['contrato'].value_counts())
```

```python
df['porte'] = df['porte'].map({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})

print("Valores da coluna 'porte' traduzidos:")
display(df['porte'].value_counts())
```

```python
print("Valores da coluna 'porte' traduzidos:")
display(df['porte'].value_counts())
```

```python
df['remoto'] = df['remoto'].map({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
})

print("Valores da coluna 'remoto' traduzidos:")
display(df['remoto'].value_counts())
```

```python
import numpy as np
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})
# calcula a média salarial e substitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

'''Calcula a mediana salarial e substitui os nulos pela mediana
'''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
print(df_salarios)
df_salarios

```
