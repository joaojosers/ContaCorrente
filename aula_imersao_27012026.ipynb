```python
import numpy as np
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})
# calcula a média salarial e substitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

'''Calcula a mediana salarial e substitui os nulos pela mediana
'''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
print(df_salarios)
df_salarios

```

```python
import plotly.express as px
```

```python
salario_medio_por_senioridade = df_limpo.groupby('senioridade')['usd'].mean().reset_index()
salario_medio_por_senioridade = salario_medio_por_senioridade.sort_values(by='usd', ascending=False)

fig = px.bar(salario_medio_por_senioridade,
             y='usd',
             x='senioridade',
             title='Média Salarial por Senioridade (Plotly)',
             labels={'senioridade': 'Nível de Senioridade', 'usd': 'Média Salarial em USD'},
             color='senioridade', # Optionally color bars by seniority
             template='plotly_white') # Use a clean template

fig.update_layout(xaxis_title='Senioridade', yaxis_title='Salário Médio em Dólar')
fig.show()
```

```python
df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']

salario_medio_ds_por_residencia = df_data_scientist.groupby('residencia')['usd'].mean().reset_index()
salario_medio_ds_por_residencia = salario_medio_ds_por_residencia.sort_values(by='usd', ascending=False)

# Mapeamento de siglas de países para nomes completos
country_mapping = {
    'US': 'United States', 'SG': 'Singapore', 'BG': 'Bulgaria', 'CH': 'Switzerland',
    'EG': 'Egypt', 'CA': 'Canada', 'MX': 'Mexico', 'NZ': 'New Zealand',
    'JP': 'Japan', 'AU': 'Australia', 'IT': 'Italy', 'DO': 'Dominican Republic',
    'DZ': 'Algeria', 'KR': 'South Korea', 'DE': 'Germany', 'GB': 'United Kingdom',
    'ES': 'Spain', 'FR': 'France', 'IE': 'Ireland', 'UA': 'Ukraine',
    'FI': 'Finland', 'NL': 'Netherlands', 'PL': 'Poland', 'BR': 'Brazil',
    'IN': 'India', 'NG': 'Nigeria', 'GR': 'Greece', 'HR': 'Croatia',
    'RO': 'Romania', 'SE': 'Sweden', 'TR': 'Turkey', 'DK': 'Denmark',
    'AE': 'United Arab Emirates', 'AT': 'Austria', 'BE': 'Belgium', 'CZ': 'Czech Republic',
    'HK': 'Hong Kong', 'PT': 'Portugal', 'RU': 'Russia', 'SK': 'Slovakia',
    'TH': 'Thailand', 'VN': 'Vietnam', 'PK': 'Pakistan', 'HU': 'Hungary',
    'LV': 'Latvia', 'RS': 'Serbia', 'SI': 'Slovenia', 'CO': 'Colombia',
    'ID': 'Indonesia', 'IQ': 'Iraq', 'IR': 'Iran', 'GH': 'Ghana',
    'CY': 'Cyprus', 'BA': 'Bosnia and Herzegovina', 'PH': 'Philippines', 'KE': 'Kenya',
    'CL': 'Chile', 'AD': 'Andorra', 'AR': 'Argentina', 'MD': 'Moldova',
    'EE': 'Estonia', 'MT': 'Malta', 'NO': 'Norway', 'MA': 'Morocco',
    'PR': 'Puerto Rico', 'CR': 'Costa Rica', 'AM': 'Armenia', 'ZA': 'South Africa',
    'AS': 'American Samoa', 'AZ': 'Azerbaijan', 'GE': 'Georgia', 'MY': 'Malaysia',
    'NP': 'Nepal', 'PE': 'Peru', 'LT': 'Lithuania', 'LU': 'Luxembourg',
    'JE': 'Jersey', 'QA': 'Qatar', 'KW': 'Kuwait', 'CN': 'China' # Added a few more common ones
}

salario_medio_ds_por_residencia['residencia_nome_completo'] = salario_medio_ds_por_residencia['residencia'].map(country_mapping)
# Para os países que não estão no mapeamento, manter a sigla original
salario_medio_ds_por_residencia['residencia_nome_completo'] = salario_medio_ds_por_residencia['residencia_nome_completo'].fillna(salario_medio_ds_por_residencia['residencia'])

fig = px.bar(salario_medio_ds_por_residencia,
             x='residencia_nome_completo', # Usar o nome completo para o eixo X
             y='usd',
             title='Média Salarial de Data Scientists por País de Residência',
             labels={'residencia_nome_completo': 'País de Residência', 'usd': 'Média Salarial Anual (USD)'},
             color='residencia_nome_completo', # Colorir barras e legenda pelo nome completo do país
             template='plotly_white')

fig.update_layout(xaxis_title='País de Residência', yaxis_title='Média Salarial em Dólar')
fig.show()
```

```python
# Gráfico de média salarial de Data Scientists por país de residência (com siglas)

df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']

salario_medio_ds_por_residencia_sigla = df_data_scientist.groupby('residencia')['usd'].mean().reset_index()
salario_medio_ds_por_residencia_sigla = salario_medio_ds_por_residencia_sigla.sort_values(by='usd', ascending=False)

fig_sigla = px.bar(salario_medio_ds_por_residencia_sigla,
             x='residencia',
             y='usd',
             title='Média Salarial de Data Scientists por País de Residência (Siglas)',
             labels={'residencia': 'País de Residência (Sigla)', 'usd': 'Média Salarial Anual (USD)'},
             color='residencia', # Color bars by country code
             template='plotly_white')

fig_sigla.update_layout(xaxis_title='País de Residência (Sigla)', yaxis_title='Média Salarial em Dólar')
fig_sigla.show()
```
